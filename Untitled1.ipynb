```python
#!/usr/bin/env python3
"""
Simulate the potassium gating variable n(t) and conductance gK(t)
under a voltage-clamp protocol using the Hodgkin-Huxley model.

The script:
1. Defines the rate functions alpha_n and beta_n for the potassium channel's
   activation gate 'n'.
2. Computes the gating variable n(t) over time for various clamp potentials.
3. Plots the resulting potassium conductance, gK(t).
"""

import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------------------------
# 1. Model Parameters
# ------------------------------------------------------------------------------
V_rest = -65.0      # Resting membrane potential (mV)
dt = 0.02           # Time step (ms)
tmax = 10.0         # End time of simulation (ms)
t_array = np.arange(0, tmax + dt, dt)

gK_max = 36.0       # Maximum K+ conductance (mS/cm^2), classic HH value
E_K = -77.0         # K+ reversal potential (mV)

# ------------------------------------------------------------------------------
# 2. Define Rate Functions alpha_n and beta_n for the 'n' gate
#    (Classic Hodgkin-Huxley at ~6.3°C)
# ------------------------------------------------------------------------------
def alpha_n(V):
    """Rate constant alpha_n (1/ms) for the K+ activation gate."""
    # Handle the singularity at V = -55 mV where the denominator is zero.
    # A small offset can be used for numerical stability if needed.
    if V == -55.0:
        return 0.1  # This is the limit as V -> -55
    return 0.01 * (V + 55.0) / (1.0 - np.exp(-(V + 55.0) / 10.0))

def beta_n(V):
    """Rate constant beta_n (1/ms) for the K+ activation gate."""
    return 0.125 * np.exp(-(V + 65.0) / 80.0)

# Vectorize the functions to work with numpy arrays
alpha_n_vec = np.vectorize(alpha_n)
beta_n_vec = np.vectorize(beta_n)

# ------------------------------------------------------------------------------
# 3. Helper Functions for Steady-State Values and Time Constants
# ------------------------------------------------------------------------------
def n_inf(V):
    """Steady-state activation of the 'n' gate."""
    return alpha_n_vec(V) / (alpha_n_vec(V) + beta_n_vec(V))

def tau_n(V):
    """Time constant for the 'n' gate (ms)."""
    return 1.0 / (alpha_n_vec(V) + beta_n_vec(V))

# ------------------------------------------------------------------------------
# 4. Initial Conditions at Rest
#    n(t=0) = n_inf(V_rest)
# ------------------------------------------------------------------------------
n_rest = n_inf(V_rest)

# ------------------------------------------------------------------------------
# 5. Voltage-Clamp Simulation and Plotting
# ------------------------------------------------------------------------------
clamp_values = np.arange(-50, 61, 10)  # Steps from -50 mV to +60 mV

# Prepare a figure for the gK(t) results
plt.figure(figsize=(10, 7))

for Vc in clamp_values:
    # Compute the steady-state value and time constant at the clamp potential Vc
    n_eq = n_inf(Vc)
    tau_n_c = tau_n(Vc)
    
    # Array to store n(t) and gK(t) over time
    n_t = np.zeros_like(t_array)
    gK_t = np.zeros_like(t_array)
    
    # Compute the gating variable at each time step
    for i, t in enumerate(t_array):
        # Exponential relaxation from the initial (rest) value to the new steady state
        n_t[i] = n_eq - (n_eq - n_rest) * np.exp(-t / tau_n_c)
        
        # Potassium conductance: gK(t) = gK_max * n^4
        gK_t[i] = gK_max * (n_t[i]**4)
    
    # Plot gK(t) for this clamp potential
    plt.plot(t_array, gK_t, label=f'Vc = {Vc} mV')

# Finalize the plot
plt.title('Potassium Conductance ($g_K$) vs. Time for Different Clamp Potentials', fontsize=16)
plt.xlabel('Time (ms)', fontsize=12)
plt.ylabel('$g_K$ (mS/cm$^2$)', fontsize=12)
plt.legend(title='Clamp Potential', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
