```python
import numpy as np
import matplotlib.pyplot as plt

# === Simulation Parameters ===
E_Na = 54.2    # Sodium reversal potential (mV)
E_K = -74.7    # Potassium reversal potential (mV)
E_l = -65.8    # Leak reversal potential (mV)
E_rest = -68   # Resting potential (mV)

G_Na_max = 30  # Max Na+ conductance (mS/cm²)
G_K_max = 12   # Max K+ conductance (mS/cm²)
G_l = 0.001    # Leak conductance (mS/cm²)
Cm = 1         # Membrane capacitance (µF/cm²)

dt = 0.05      # Time step (ms)
T_max = 5      # Total simulation time (ms)
time = np.arange(0, T_max, dt)

# === Drug Options: "control", "lidocaine", "TTX", "TEA" ===
drug = "TTX"

# === Stimulus ===
I_stim = np.zeros_like(time)
I_stim[time < 0.3] = 75  # μA/cm² pulse for first 0.3 ms

# === Initial Conditions ===
Vm = E_rest
n, m, h = 0.3, 0.0665, 0.6

# === Functions to Prevent Overflow ===
def exp_clip(x):
    return np.exp(np.clip(x, -100, 100))

# === Rate Constants ===
def alpha_n(V): d = 10 - V; return 0.01 * d / (exp_clip(d / 10) - 1 + 1e-12)
def beta_n(V): return 0.125 * exp_clip(-V / 80)
def alpha_m(V): d = 25 - V; return 0.1 * d / (exp_clip(d / 10) - 1 + 1e-12)
def beta_m(V): return 4 * exp_clip(-V / 18)
def alpha_h(V): return 0.07 * exp_clip(-V / 20)
def beta_h(V): return 1 / (exp_clip((30 - V) / 10) + 1)

# === Storage Lists ===
Vm_values, n_values, m_values, h_values = [], [], [], []

# === Simulation Loop ===
for t_idx, t in enumerate(time):
    # Update gating variables
    a_n, b_n = alpha_n(Vm), beta_n(Vm)
    a_m, b_m = alpha_m(Vm), beta_m(Vm)
    a_h, b_h = alpha_h(Vm), beta_h(Vm)

    n += dt * (a_n * (1 - n) - b_n * n)
    m += dt * (a_m * (1 - m) - b_m * m)
    h += dt * (a_h * (1 - h) - b_h * h)

    # Compute conductances
    G_K = G_K_max * n**4
    G_Na = G_Na_max * m**3 * h

    # === Drug Effects ===
    if drug == "lidocaine":
        G_Na *= 0.5  # 50% block
    elif drug == "TTX":
        G_Na = 0     # Full Na+ block
    elif drug == "TEA":
        G_K *= 0.3   # 70% K+ block

    # Compute ionic currents
    I_K = G_K * (Vm - E_K)
    I_Na = G_Na * (Vm - E_Na)
    I_l = G_l * (Vm - E_l)
    I_total = I_stim[t_idx] - (I_Na + I_K + I_l)

    # Update membrane potential
    Vm += dt * I_total / Cm
    Vm = np.clip(Vm, -150, 100)

    # Store results
    Vm_values.append(Vm)
    n_values.append(n)
    m_values.append(m)
    h_values.append(h)

# === Plot Results ===
plt.figure(figsize=(10, 6))
plt.plot(time, Vm_values, label=f"Membrane Potential (drug: {drug})", color='black')
plt.axhline(y=E_rest, linestyle="--", color="gray", label="Resting Potential")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.title("Neuron Action Potential Simulation with Drug Effects")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

# === Simulation Parameters ===
E_Na = 54.2
E_K = -74.7
E_l = -65.8
E_rest = -68

G_Na_max = 30
G_K_max = 12
G_l = 0.001
Cm = 1

# === Time Parameters ===
dt = 0.05
T_max = 5

def exp_clip(x):
    return np.exp(np.clip(x, -100, 100))

def alpha_n(V): d = 10 - V; return 0.01 * d / (exp_clip(d / 10) - 1 + 1e-12)
def beta_n(V): return 0.125 * exp_clip(-V / 80)
def alpha_m(V): d = 25 - V; return 0.1 * d / (exp_clip(d / 10) - 1 + 1e-12)
def beta_m(V): return 4 * exp_clip(-V / 18)
def alpha_h(V): return 0.07 * exp_clip(-V / 20)
def beta_h(V): return 1 / (exp_clip((30 - V) / 10) + 1)

def simulate(drug):
    time = np.arange(0, T_max, dt)
    I_stim = np.zeros_like(time)
    I_stim[time < 0.3] = 75

    Vm = E_rest
    n, m, h = 0.3, 0.0665, 0.6

    Vm_values = []

    for t_idx, t in enumerate(time):
        a_n, b_n = alpha_n(Vm), beta_n(Vm)
        a_m, b_m = alpha_m(Vm), beta_m(Vm)
        a_h, b_h = alpha_h(Vm), beta_h(Vm)

        n += dt * (a_n * (1 - n) - b_n * n)
        m += dt * (a_m * (1 - m) - b_m * m)
        h += dt * (a_h * (1 - h) - b_h * h)

        G_K = G_K_max * n**4
        G_Na = G_Na_max * m**3 * h

        # Drug Effects
        if drug == "Lidocaine":
            G_Na *= 0.5
        elif drug == "TTX":
            G_Na = 0
        elif drug == "TEA":
            G_K *= 0.3

        I_K = G_K * (Vm - E_K)
        I_Na = G_Na * (Vm - E_Na)
        I_l = G_l * (Vm - E_l)
        I_ext = I_stim[t_idx]

        Vm += dt * (I_ext - (I_Na + I_K + I_l)) / Cm
        Vm = np.clip(Vm, -150, 100)

        Vm_values.append(Vm)

    return time, Vm_values

# === Streamlit UI ===
st.title("Neuron Drug Effect Simulator")

drug_choice = st.selectbox("Select a drug to simulate:", ["Control", "Lidocaine", "TTX", "TEA"])

if st.button("Run Simulation"):
    time, Vm_values = simulate(drug_choice)

    st.subheader(f"Membrane Potential over Time ({drug_choice})")
    fig, ax = plt.subplots()
    ax.plot(time, Vm_values, label=f"Vm with {drug_choice}", color='black')
    ax.axhline(y=E_rest, linestyle="--", color="gray", label="Resting Potential")
    ax.set_xlabel("Time (ms)")
    ax.set_ylabel("Membrane Potential (mV)")
    ax.set_title("Neuron Action Potential Simulation")
    ax.legend()
    ax.grid(True)
    st.pyplot(fig)

    st.success("Simulation complete!")
```
